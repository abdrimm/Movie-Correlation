```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv('/Users/macbook/Downloads/movies.csv')
```

```python
#looking at the data
df.head()
```

```python
# looking for any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#dropping missing data
#df = df.dropna()
```

```python
df.dtypes
```

```python
df = df.fillna(0) 
#fixing error for next step
```

```python
#change data type of columns
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
#budget correlation
#scatter plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs gross earnings')
plt.xlabel('Budget earnings')
plt.ylabel('Gross for film')
plt.show()
```

```python
#plot budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df)
plt.title('Budget vs gross earnings')
plt.xlabel('Budget earnings')
plt.ylabel('Gross for film')
```

```python
#looking at correlation
df.corr(method = 'pearson') #pearson, kendall, spearman
```

```python
#high correlation between budget and gross
#visualize this table
correlarion_matrix = df.corr(method = 'pearson')
sns.heatmap(correlarion_matrix, annot=True)
plt.title('Correlation matrix for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
#company correlation
#numerize company names

df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python

correlarion_matrix = df_numerized.corr(method = 'pearson')
sns.heatmap(correlarion_matrix, annot=True)
plt.title('Correlation matrix for all features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
df_numerized.corr(method = 'pearson')
```

```python
correlation_matr = df_numerized.corr(method = 'pearson')
corr_pairs = correlation_matr.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_correlation = sorted_pairs[(sorted_pairs)>0.5]
high_correlation
```
